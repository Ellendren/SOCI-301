1. Choose TrafDths09 (the rate of traffic fatalities in 2009) and TrafDths90 (the rate of traffic fatalities in 1990) variables and produce the following statistics: mean, median, standard deviation, and range.
---

```python
# save dired neames as variables
TrafDths09 = "TrafDths09"
TrafDths90 = "TrafDths90"

# extract TrafDths90 and TrafDths09 colums
traf_dths_df = states_df[[TrafDths09, TrafDths90]]

# get the desired stats
list_index_labels = ['mean', 'median', 'standard deviation', 'range']

mean_09 = traf_dths_df[TrafDths09].mean()

meadian_09 = traf_dths_df[TrafDths09].median()

std_09 = traf_dths_df[TrafDths09].std()

range_09 = traf_dths_df[TrafDths09].max() - traf_dths_df[TrafDths09].min()

mean_90 = traf_dths_df[TrafDths90].mean()

meadian_90 = traf_dths_df[TrafDths90].median()

std_90 = traf_dths_df[TrafDths90].std()

range_90 = traf_dths_df[TrafDths90].max() - traf_dths_df[TrafDths90].min()

list_of_stats = [(mean_09, mean_90),
                 (meadian_09, meadian_90),
                 (std_09, std_90),
                 (range_09, range_90)]

#condense into data frame
pd_traf_stats = pd.DataFrame(data=list_of_stats, columns=[TrafDths09, TrafDths90], index=list_index_labels)
display(pd_traf_stats)
```
